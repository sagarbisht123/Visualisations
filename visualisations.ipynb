## Line Chart 
The line chart is one of the most simplest and widely used visualisation techniques. A line chart displays a series of data connected with the line. We can change the colour, shape and customize the markers etc.

```python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

Lets visulaise this data with the line chart. For this we will use the <code>plt.plot</code> function.

```python
plt.plot(yield_apples);
```

## Customising The X-Axis
As we can see in the graph above the X-axis is actually consist of the indexes.
Here lets pass another series for the X-axis.


```python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
years=[2010,2011,2012,2013,2014,2015]
plt.plot(years,yield_apples);
```

```python
#We can add labels to the X-axis and Y-axis.
plt.plot(years, yield_apples);
plt.xlabel("year ---------->")
plt.ylabel("Yield of apples[tons per hectare] ----->")
```

# Histogram

```python
#Load dataset into data frame
flowers_df=sns.load_dataset("iris")
```

```python
flowers_df.sepal_width
```

```python
plt.title("Distribution Of Sepal Width")
plt.hist(flowers_df.sepal_width)
```

# Controlling The size of bins


We can control the size of bins in several ways.
In the first method we can tell them the number of bins

```python
plt.title("Distribution of Sepal Length")
plt.hist(flowers_df.sepal_width,bins=5)
```

```python
#Controlling the size of bins by giving them an equal width by arange method
import numpy as np
plt.title("Distribution of sepal width")
plt.hist(flowers_df.sepal_width, bins=np.arange(2,5,0.25))
```

```python
#For the bins of the unequal length we can pass a list 
plt.title("Distribution Of sepal width")
plt.hist(flowers_df.sepal_width, bins=[1,2.5,3.5,4.5])
```

# Multiple Histograms 
We can multiple histogram for suppose <code>flower_df</code>. One way to do it in a good way is to decrease visibility by the <code>alpha</code>

lets plot a multiple histogram for the sepal width of three different species of flowers 

```python
setosa_df=flowers_df[flowers_df.species=="setosa"]
versicolor_df=flowers_df[flowers_df.species=="versicolor"]
virginica_df=flowers_df[flowers_df.species=="virginica"]
```

```python
plt.hist(setosa_df.sepal_width,alpha=0.4, bins=5)
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=5)
```

What if I want to stack them one over another? For this we can use <code>stacked</code>.

```python
plt.title("Distribution of sepal width")
plt.hist([setosa_df.sepal_width, virginica_df.sepal_width,versicolor_df.sepal_width], bins=np.arange(2,5,0.25), stacked=True)
plt.legend(["Setosa","Virginica","Versicolor"])
```

# Heatmap

A heatmap is used to visulaise a 2D-matrix like a matrix using colors. The best way to understand is by looking an example of <code>flights_df</code>.
It gives us the data about the flights in a month at different airports.

```python
flights=sns.load_dataset("flights")
flights_df=flights.pivot(index="month",columns="year",values="passengers")
flights_df
```

```python
plt.title("Passengers------>")
sns.heatmap(flights_df)
```

We can also display the values in each block by <code>annot</code> and we can display the blue pallete as <code>cmap</code>.

```python
plt.title("Passengers---->")
sns.heatmap(flights_df,fmt="d", annot=True,cmap="Blues")
```
